# **1.Handling Missing Value-Ignore tuple**

```python
#Menghapus baris kosong embarked dan mereplace data baru ke dataframe
#dropna untuk menghapus value

df.dropna(subset=['Embarked'], inplace = True) #Jadi kosong karna dihapus/direplace
```

```python
#Cek kembali keberadaan nilai kosong
df.isnull().sum()
```

# **2.Handling Missing Value-Fill using attibute mean, median or modus**

```python
#Kolom Age, kita isikan dengan nilai rata-rata dari data yang lain
#Kalo tipe data float menggunakan mean

rata_umur = df['Age'].mean() #Langkah 1 : hitung rata-rata dari fitur age
df['Age'] = df['Age'].fillna(rata_umur) #Langkah 2 : replace/isikan data mean ke field yang kosong
print('Jumlah data null pada kolom Age :', df['Age'].isna().sum()) #Langkah 3
```

```python
#Kalo data null banyak boleh diisi atau di drop
#Cara mengisi data cabin yang tipe datanya string menggunakan modus(nilai yang sering muncul)
#Langkah 1

modus = df['Cabin'].mode()[0]
print(modus)
```

```python
#Buktikan bahwa nilai variabel 'modus'sudah benar
#lakukan grouping berdasarkan value cabin kemudian menghitungnya

df.groupby('Cabin')['Cabin'].count().sort_values(ascending=False) #Ascending kecil-besar, descending besar-kecil
```

```python
df['Cabin'] = df['Cabin'].fillna(modus) #Langkah 2
df['Cabin'].isna().sum() #Langkah 3
```

```python
df.isnull().sum()
```

```python
#Menghapus kolom tertentu
#Kolom Fare, kita abaikan maka kita hapus saja kolom dan baris tertentu
print('Kolom Awal : ')
print(df.columns)

df.drop('Fare', axis=1, inplace=True) #drop untuk menghapus #dropna untuk menghapus value
#axis=0 Baris
#Axis=1 Kolom

print('\n\nKolom Setelah Fare dihapus : ')
print(df.columns)
```
