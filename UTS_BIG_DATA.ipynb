# 6. Transform RDD

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

```python
rdd = sc.parallelize([('Anggista', "20113690"),('Teddy',"20113668"),('Dhea',"20113725"),('Zalza',"20113682")])

rdd_reduced = rdd.reduceByKey(lambda x, y: x+y)

for num in rdd_reduced.collect():
    print("Nama {} dengan NIM {}".format(num[0], num[1]))
```

```python
rdd_reduced.collect()
```

```python
rdd_reduced_sort = rdd_reduced.sortByKey(ascending=True) #Kalo false jadi kebalikannya

for num in rdd_reduced_sort.collect():
  print("Nama {} memiliki NIM {}".format(num[0], num[1]))
```

```python
total = rdd.countByKey()

for k, v in total.items():
    print("Nama", k, "memiliki", v, "NIM")
```
